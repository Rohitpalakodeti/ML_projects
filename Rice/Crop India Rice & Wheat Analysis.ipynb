```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df=pd.read_csv('Final_crop_data (1) (1).csv')
```

```python
df.drop('Unnamed: 15',axis=1,inplace=True)
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df=df[df['sunhours']>0]

df.info()
```

```python
def Season(n):
    
    if n=='Autumn     ':
        return 0
    elif n=='Kharif     ':
        return 1
    elif n=='Rabi       ':
        return 2
    elif n=='Whole Year ':
        return 3
    elif n=='Summer     ':
        return 4
    elif n=='Winter     ':
        return 5
    

df['Season_code']=df['Season'].apply(Season)
    
```

```python
df.head()
```

```python
print(df['State'].unique())
print(df['District'].unique())
```

```python
districts=[]
districts=df['District'].unique()
def district_code(n):
    
    for i in range(len(districts)):
        
        if districts[i]==n:
            
            return i
        
df['District_code']=df['District'].apply(district_code)
        
    
```

```python
states=[]

states=df['State'].unique()

def state_code(n):
    
    for i in range(len(states)):
        
        if states[i]==n:
            return i
        
        
df['State_code']=df['State'].apply(state_code)
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression,Lasso,Ridge
from sklearn.svm import SVC,SVR
from sklearn.cross_validation import train_test_split

X=df[['Area','Avg_Temperature', 'rainfall', 'Avg_windspeed', 'Avg_pressure',
      'Avg_humidity', 'Avg_uvindex', 'sunhours', 'Season_code',
       'District_code', 'State_code']]

y=df['Production']


x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=101)

lin=LinearRegression()

lin.fit(x_train,y_train)

print('Linear Regression : {}'.format(lin.score(x_test,y_test)))

svm=SVR()

svm.fit(x_train,y_train)

print('Support Vector Regressor : {}'.format(svm.score(x_test,y_test)))
```
